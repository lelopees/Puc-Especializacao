```python
from sklearn.datasets import load_digits
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict
import scikitplot as skplt
```

```python
X, y = load_digits(return_X_y=True)
```

```python
X
```

```python
y
```

```python
rf = RandomForestClassifier(n_estimators=5, max_depth=5, random_state=1)
```

```python
predictions = cross_val_predict(rf,X,y)
```

```python
#lateral esquerda = valores reais
#menu inferior = classes preditas
skplt.metrics.plot_confusion_matrix(y, predictions, normalize=True)
```

Além disso, a matriz de confusão permite identificar os erros, sendo possível saber qual classe foi predita em uma estimativa errada. Por exemplo, o número 8 possui uma baixa taxa de acertos na predição contendo somente 59%, sendo que, desses erros, a classe mais predita de forma errada é a 1 com 14% de predições erradas, seguida da classe 9 com apenas 6%. Esses dados podem ser lidos na linha (true label) que representa a classe 8. Outro número que é bastante confundido pelo classificado é o número 3, sendo confundido com o número 5. O número que possui a maior taxa de acertos é o 0 com 95% de acertos, seguido pelo 6 com 93% de acertos.
