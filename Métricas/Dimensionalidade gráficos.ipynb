```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets      import load_iris
from matplotlib.colors     import ListedColormap
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.manifold      import TSNE
```

```python
dataset = load_iris()
features = ['sepal_length','sepal_width','petal_length','petal_width']
target = 'species'
iris = pd.DataFrame(dataset.data, columns = features)
iris[target] = dataset.target
```

```python
sns.set()
sns.set(rc={"figure.figsize": (10,8)})
PALETTE = sns.color_palette('deep', n_colors=3)
CMAP = ListedColormap(PALETTE.as_hex())
sns.set_style("darkgrid")
```

```python
plt.scatter(iris['sepal_length'],
           iris['sepal_width'], c=iris['species'],
           cmap=CMAP, s=70)
plt.title('Dados originais', fontsize=20, y=1.03)
plt.xlabel('Sepal length', fontsize=16)
plt.xlabel('Sepal width', fontsize=16)
plt.show()
```

```python
caracteristicas = StandardScaler().fit_transform(iris[features])

points = pca.fit_transform(caracteristicas)

pca = PCA(n_components=2)

plt.scatter(points[:,0], points[:,1],
           c=iris['species'], cmap=CMAP, s=70)

plt.title('Dados usando PCA', fontsize=20, y=1.03)
plt.xlabel('PCA-1', fontsize=16)
plt.ylabel('PCA 2', fontsize=16)
plt.show()
```

```python
tsne = TSNE(n_components=2, n_iter=1000, random_state=0)

caracteristicas_tsne = tsne.fit_transform(caracteristicas)

points = tsne.fit_transform(caracteristicas_tsne)

plt.scatter(points[:,0], points[:,1], c=iris['species'], cmap=CMAP, s=70)

plt.title('Dados usando t-SNE', fontsize=20, y=1.03)
plt.xlabel('t-SNE 1', fontsize=16)
plt.ylabel('t-SNE 2', fontsize=16)
plt.show()
```
