```python
import matplotlib.pyplot as plt
import scikitplot as skplt
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()
X, y = data.data, data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_probas = nb.predict_proba(X_test)
```

```python
skplt.metrics.plot_precision_recall(y_test, y_probas)
plt.show()
```

```python
skplt.metrics.plot_roc(y_test, y_probas)
```

A curva ROC apresentou um valor de 99% tanto para a classe 0 como para a classe 1 (maligno/no maligno), sendo considerado um valor extremamente bom, pois o modelo gerado possui uma alta qualidade para os dados em que foi testado.
