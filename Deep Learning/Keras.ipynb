```python
from numpy import loadtxt
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

```python
# carregando dados de Pima Indians Diabetes
df = loadtxt(
    'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv', delimiter=',')
```

```python
x = df[:, 0:8]
y = df[:,8]
```

```python
#Primeiramente, precisamos garantir que a camada de entrada tem a quantidade correta de inputs de entrada
# Isso pode ser especificado no Keras utilizando o argumento input_dim e o ajustando para 8, nosso número de entradas

#iniciando modelo de sequencial
model = Sequential()
# incializar a primeira camada, com 12 neuronios, 8 entradas e a função relu
model.add(Dense(12, input_dim=8, activation='relu'))
# segunda camada com 8 neuronios e função relu
model.add(Dense(8, activation='relu'))
# ultima camada com um neuronio e função sigmoid
model.add(Dense(1,activation='sigmoid'))
```

```python
# função de perda, otimizador e métrica de avaliação
# adam utiliza o gradiente descente
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=150, batch_size=10)
```

```python
_, accuracy = model.evaluate(x,y)
print('Acurácia: %.2f' % (accuracy*100))
```
