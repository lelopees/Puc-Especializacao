```python
import pandas as pd
import numpy as np
from category_encoders.one_hot import OneHotEncoder

import warnings
warnings.filterwarnings('ignore')
```

```python
dados = {'Country': ['India', 'US', 'Japan', 'US', 'Japan'],
        'Ativo': ['Não', 'Sim', 'Sim', 'Não', 'Sim'],
        'Age' : [44, 34, 46, 35, 23],
        'Salary' : [ 72000, 65000, 98000, 45000, 34000]}

df = pd.DataFrame(dados) 
df
```

```python
df_catOne = df.select_dtypes(include='object')
```

```python
# variavel categorica
df_catOne
```

```python
df_catOne.columns
```

```python
one_hot = OneHotEncoder(cols=df_catOne.columns)
```

```python
# transformando os dados categoricos em numericos
df_one = one_hot.fit_transform(df_catOne)
```

```python
df_one
```

```python
df = pd.concat([df, df_one], axis=1)
```

```python
# incluido os campos novos transformados em numericos no dataframe
df
```

```python
# exclusão do campo antigo Country
df = df.drop(['Country', 'Ativo'], axis = 1)
```
