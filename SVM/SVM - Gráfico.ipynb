```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets     import make_circles
from mpl_toolkits.mplot3d import Axes3D
```

```python
# caracteristicas e rotulos
caracteristicas, c = make_circles(n_samples=500, noise=0.09)
```

```python
#define as cores dos pontos no grÃ¡fico
rgb = np.array(['r','g'])
rgb
```

```python
# cria e exibe o grÃ¡fico do tipo Scatter plot
plt.scatter(caracteristicas[:,0], caracteristicas[:,1], color=rgb[c])
plt.show()
```

```python
#grÃ¡fico com visualizaÃ§Ã£o 3D
fig = plt.figure(figsize=(18,5))
# adiciona a visualizaÃ§Ã£ em 3 dimensÃµes
ax = fig.add_subplot(111,projection='3d')
z = caracteristicas[:,0]**2+caracteristicas[:,1]**2
# cria o grÃ¡fico e exibe
ax.scatter(caracteristicas[:,0], caracteristicas[:,1],z,color=rgb[c])
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```
