```python
import matplotlib
import pandas as pd
import numpy as np
from sklearn import svm, datasets
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
print(iris.data[0:5])
```

```python
print(iris.feature_names)
```

```python
print(iris.target_names)
```

```python
X = iris.data[:,:2]
y = iris.target
```

```python
colors = ['red','green','blue']

for color, i, target in zip(colors, [0,1,2], iris.target_names):
    plt.scatter(X[y==i,0], X[y==i,1], color=color, label=target)
    
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.legend(loc='best', shadow=False, scatterpoints=1)

plt.title('Scatter plot: Sepal width vc Sepal length')
plt.show()
```

```python
# parametro de regularização do SVM
C = 1

#cria o modelo usando o SVM

clf = svm.SVC(kernel='linear', C=C).fit(X,y)
title = 'SVC com kernel linear'

# comandos para criar as marcações de separação
x_min, x_max = X[:,0].min() -1, X[:,0].max() + 1 
y_min, y_max = X[:,1].min() -1, X[:,1].max() + 1

h = (x_max / x_min) / 100

# cria uma grade retangular a partir de uma matriz de valores
xx,yy = np.meshgrid(np.arange(x_min, x_max,h), np.arange(y_min, y_max, h))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
# nova forma à matriz sem alterar os dados
Z = Z.reshape(xx.shape)

plt.contourf(xx,yy, Z, cmap=plt.cm.Accent, alpha=0.8)

#adicionar os pontos ao gráfico colorido criado
colors = ['red','green','blue']


for color, i, target in zip(colors,[0,1,2], iris.target_names):
    plt.scatter(X[y==i,0], X[y==i,1], color=color, label=target)
    
    
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title(title)
plt.legend(shadow=False, scatterpoints=1)

predictions = clf.predict(X)
print(np.unique(predictions, return_counts=True))

# 50 setosa, 53 versicolor e 47 virginica
```
