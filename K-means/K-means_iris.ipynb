```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

warnings.filterwarnings('ignore')
```

```python
iris = load_iris() 
X = iris.data
y = iris.target
```

```python
wcss = [] 

for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('NÃºmero de Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
Kmeans_clusters = KMeans(n_clusters=3, random_state=10)
cluster_labels = Kmeans_clusters.fit_predict(X)
```

```python
print(f'Total de dados: {len(y)} ')
print(f'Total de clusters "identificados" como corretamente {(y == cluster_labels).sum()}')
```

```python
plt.scatter(X[cluster_labels == 0, 0], X[cluster_labels == 0, 1], s=100, c= 'green', label = 'Setosa')
plt.scatter(X[cluster_labels == 1, 0], X[cluster_labels == 1, 1], s=100, c= 'red', label = 'Versicolour')
plt.scatter(X[cluster_labels == 2, 0], X[cluster_labels == 2, 1], s=100, c='orange', label='Virginica')
plt.gcf().set_size_inches(8,6)
plt.legend()
plt.show()
```
