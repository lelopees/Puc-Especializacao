```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
# noise = indica a portcentagem de ruido que ter√° nos dados
X, y = make_moons(n_samples=200, noise=0.05, random_state=23)
```

```python
# Defini√ß√£o de atributos
# marker = tipo de ponto do gr√°fico (o,s,p)
# c=cor
# s=tamanho do ponto no gr√°fico
plt.scatter(X[y==0,0], X[y==0,1],c='green', marker='s', s=40, label='cluster 1')
plt.scatter(X[y==1,0], X[y==1,1], c='red', marker='o', s=40, label='cluster 2')
plt.legend()
plt.show()
```

```python
# eps = como ser√° feita c clusteriza√ß√£o, espeficia a distancia maxima ente os elementos
# min_samples = qauntidade de amostrar mininas para serem consideradas na vizinhan√ßa para ser considerado como core point
db = DBSCAN(eps=0.2, min_samples=5, metric="euclidean")
```

```python
# realizada a clusteriza√ß√£o nos dados armazenados em x e armana na variavel y_db
y_db = db.fit_predict(X)
```

```python
plt.scatter(X[y_db == 0, 0], X[y_db == 0, 1], c="yellow", marker="o", s=40, label="cluster 1")
plt.scatter(X[y_db == 1, 0], X[y_db == 1, 1], c="green",  marker="s", s=40, label="cluster 2")
plt.legend()
plt.show()
```
